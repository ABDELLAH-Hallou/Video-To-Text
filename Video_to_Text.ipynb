```python
video = mp.VideoFileClip(output_file_path+videoName)
video.audio.write_audiofile(output_file_path+audioName)
```

```python
sound = AudioSegment.from_mp3(output_file_path+audioName)
chunks_length = 10000
chunks = make_chunks(sound, chunks_length)
```

```python
def s2t(filename):
  r = sr.Recognizer()
  with sr.AudioFile(filename) as source:
      r.adjust_for_ambient_noise(source, duration=0.2)
      audio = r.record(source)
  text = r.recognize_google(audio)
  return text
```

```python
text = ""
for i, chunk in enumerate(chunks):
  chunkName = "chunk_"+str(i)+".wav"
  chunk.export(output_file_path+chunkName, format="wav")
  filename = output_file_path+chunkName
  text+=s2t(filename)
```
