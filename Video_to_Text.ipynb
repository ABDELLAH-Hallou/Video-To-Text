```python
from IPython.display import YouTubeVideo
from google.colab import drive
import os, sys
from google.colab import files
from pytube import YouTube
import moviepy.editor as mp
import speech_recognition as sr
from pydub import AudioSegment
```

```python
vId= "QhFMDUH-Le0"
link = "https://www.youtube.com/watch?v="+vId
YouTubeVideo(vId)
```

```python
drive.mount('/content/drive')
videoName = 'video.mp4'
audioName = "audio.mp3"
exportedAudioName = "audio.wav"
file_path = "/content/"+videoName
output_file_path = '/content/drive/MyDrive/work folder/'
```

```python
youtube = YouTube("https://youtu.be/"+vId)
print(youtube.title)
```

```python
video = youtube.streams.get_highest_resolution()
video.download(filename=videoName)
```

```python
!cp "$file_path" "$output_file_path"
os.remove(file_path)
```

```python
video = mp.VideoFileClip(output_file_path+videoName)
video.audio.write_audiofile(output_file_path+audioName)
```

```python
sound = AudioSegment.from_mp3(output_file_path+audioName)
audio_10s = sound[:10000]
audio_10s.export(output_file_path+exportedAudioName, format="wav")
filename = output_file_path+exportedAudioName
```

```python
r = sr.Recognizer()
with sr.AudioFile(filename) as source:
    r.adjust_for_ambient_noise(source, duration=0.2)
    audio = r.record(source)
MyText = r.recognize_google(audio)
print(MyText)
```
